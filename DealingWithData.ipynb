```python
# Downloading data and Show data

import pandas as pd
import numpy as np

data = pd.read_csv('https://raw.githubusercontent.com/realpython/python-data-cleaning/master/Datasets/BL-Flickr-Images-Book.csv')
print(type(data))
data.head()
```

```python
# Excluding irrelevant columns from data
# using python's drop() to filter or delete columns from original data
to_drop = ['Edition Statement',
'Corporate Author',
'Corporate Contributors',
'Former owner',
'Engraver',
'Contributors',
'Issuance type',
'Shelfmarks']
data.drop(to_drop, inplace=True, axis=1)
data.head()
```

```python
# Since we have identifier which is unique in whole data, therefore we use this identifier as index rather 
# than python's default index

data['Identifier'].is_unique
data= data.set_index('Identifier')
data.head()
```

```python
# We can access our dataframe using below command
data.loc[206]

# We can get total number of columns 
data.get_dtype_counts()
```

```python
# dataframe.loc is only use for the index

data['Date of Publication']
# A particular book can have only one date of publication
```

```python
# we will use this regex to clean double date patterns
extract = data['Date of Publication'].str.extract(r'^(\d{4})', expand=False)
extract.head() , data.loc[206: ,'Date of Publication'].head()
```

```python
data['Date of Publication']= pd.to_numeric(extract)
data['Date of Publication'].dtype
```

```python
data['Date of Publication']
```

```python
# We see that Place of Publication has some unnecessary information

data['Place of Publication'].head(10)
```

```python
# lets look at some other data
print(data.loc[4157862])
print("************************************************************************************")
print(data.loc[4159587])

# these are two books published in same place but one is surrounded by hyphens while the other is not
```

```python
# to clean this we can .str.contains()
publication = data['Place of Publication']
london = publication.str.contains('London')
oxford = publication.str.contains('Oxford')
```

```python
print(london[:5])
print(oxford[:5])
```

```python
# we can combine them using np.where

data['Place of Publication'] = np.where(london, 'London',
                                      np.where(oxford, 'Oxford',
                                               publication.str.replace('-', ' ')))
```

```python
# Example 2, Cleaning Dataframe using pandas.applymap(), 
# it works as python map() which take 2 parameters, one as function and the other data

university_data = []
with open('Dealing_with_data/university_data.txt') as file:
    for line in file:
        if '[edit]' in line:
            state= line
        else:
            university_data.append((state, line))
university_data[:5]
```

```python
# Making the uniersity data as Pandas dataframe and then Naming the Columns as State and Region
town_dataframe = pd.DataFrame(university_data,columns=['State', 'Region'])
```

```python
town_dataframe.head()
```

```python
# Applying panda's Dataframe .applymap() function which apply on tuples of Dataframe
# to clean Column State from edit and Region from unnecessary information.

def get_citystate(item):
    if ' (' in item:
        return item[:item.find(' (')]
    elif '[' in item:
        return item[:item.find('[')]
    else:
        return item
town_dataframe = town_dataframe.applymap(get_citystate)
town_dataframe.head(10)
```

```python
# Example 3, Renaming Columns and Skipping rows 
olumpics_data = pd.read_csv('Dealing_with_data/olumpics.csv' ,header=1)
olumpics_data.head()
```

```python
new_name = {
    'Unnamed: 0': 'Country',
    '? Summer':'Summer Olympics',
    '01 !':'Gold',
    '02 !':'Silver',
    '03 !':'Bronze',
    '? Winter': 'Winter Olympics',
    '01 !.1': 'Gold.1',
    '02 !.1': 'Silver.1',
    '03 !.1':'Bronze:1',
    '? Games':'Number of Games',
    '01 !.2': 'Gold.2',
    '02 !.2': 'Silver.2',
    '03 !.2': 'Bronze.2',
}
```

```python
# Renaming column names using pandas .rename() function
olumpics_data.rename(columns= new_name,
                     inplace=True) # inplace to True tells that changes to be made directly to objects
```
